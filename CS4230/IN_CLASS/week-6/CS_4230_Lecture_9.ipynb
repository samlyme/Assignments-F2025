# **Generate ER Graph**

Tasks:

- Increase ùëù. When does it connect?

- Check average path length of the largest component.

- How clustered is it?

```python
import networkx as nx
import matplotlib.pyplot as plt

n = 200
p = 0.03
G = nx.erdos_renyi_graph(n, p)

deg = [d for _, d in G.degree()]
print("avg degree:", sum(deg)/n)
print("components:", nx.number_connected_components(G))

#largest connected component of graph G
LCC = G.subgraph(max(nx.connected_components(G), key=len)).copy()

nx.draw(G, node_size=20, with_labels=False)
plt.show()
```

# **Generate WS Graph**

Tasks:
- test ùõΩ ‚àà {0, 0.01, 0.05, 0.1, 1}.
- Plot ùê∂ and ùêø.

```python
n, k, beta = 200, 6, 0.05
WS = nx.watts_strogatz_graph(n, k, beta)

print("avg clustering:", nx.average_clustering(WS))
LCC = WS.subgraph(max(nx.connected_components(WS), key=len)).copy()
print("avg shortest path:", nx.average_shortest_path_length(LCC))

nx.draw(WS, node_size=20, with_labels=False)
plt.show()
```

# **Generate BA Graph**

Task: Compare BA hubs to ER and WS. Which network is most realistic?

```python
n, m = 200, 2
BA = nx.barabasi_albert_graph(n, m)

deg = [d for _, d in BA.degree()]
print("avg degree:", sum(deg)/n, "max degree:", max(deg))

nx.draw(BA, node_size=20, with_labels=False)
plt.show()
```
